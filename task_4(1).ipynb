```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
ad=pd.read_csv(r"C:\Users\Aashi\Downloads\USvideos.csv")
```

```python
ad.head()
```

```python
ad['trending_date'] = pd.to_datetime(ad['trending_date'], format='%y.%d.%m')
ad.head(4)
```

```python

ad["publish_time"] = pd.to_datetime(ad["publish_time"])
print(ad.head(2))
```

```python
ad["publish_month"] = ad["publish_time"].dt.month
ad["publish_year"] = ad["publish_time"].dt.year
ad['publish_hour'] = ad['publish_time'].dt.hour
ad.head(4)
```

```python
print(sorted(ad["category_id"].unique()))
```

```python
ad['category_name']=np.nan
ad.loc[(ad["category_id"]==1),"category_name"]='Film and Animation'
ad.loc[(ad["category_id"]==2),"category_name"]='Authors and Vehicles'
ad.loc[(ad["category_id"]==10),"category_name"]='Music'
ad.loc[(ad["category_id"]==15),"category_name"]='Pets and Animals'
ad.loc[(ad["category_id"]==17),"category_name"]='Sports'
ad.loc[(ad["category_id"]==19),"category_name"]='Travel and Events'
ad.loc[(ad["category_id"]==20),"category_name"]='Gaming'
ad.loc[(ad["category_id"]==22),"category_name"]='People and Blogs'
ad.loc[(ad["category_id"]==23),"category_name"]='Comedy'
ad.loc[(ad["category_id"]==24),"category_name"]='Enterntainment'
ad.loc[(ad["category_id"]==25),"category_name"]='News and Politics'
ad.loc[(ad["category_id"]==26),"category_name"]='How to and Style'
ad.loc[(ad["category_id"]==27),"category_name"]='Science an technology'
ad.loc[(ad["category_id"]==28),"category_name"]='Non-Profits'
ad.loc[(ad["category_id"]==29),"category_name"]='Movies'
ad.loc[(ad["category_id"]==43),"category_name"]='Shows'
ad.head()
```

```python
ad['year']=ad['publish_time'].dt.year
yearly_counts=ad.groupby('year')['video_id'].count()
#bar chart
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total publish count', title='Total Publish Video per year')
plt.show()
```

```python
yearly_views=ad.groupby('year')['views'].sum()
yearly_views.plot(kind='bar', xlabel='year', ylabel='Toatl Views', title='Total views per year')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
category_views=ad.groupby('category_name')['views'].sum().reset_index()
top_categories=category_views.sort_values(by='views', ascending=False).head(5)
plt.bar(top_categories['category_name'],top_categories['views'])
plt.xlabel('Category Name',fontsize=12)
plt.ylabel('Toatl Views',fontsize=12)
plt.title('Top 5 Categories', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='category_name', data=ad,order=ad['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video count by category')
plt.show()
```

```python
videos_per_hour=ad['publish_hour'].value_counts().sort_index()
plt.figure(figsize=(12,6))
sns.barplot(x=videos_per_hour.index, y=videos_per_hour.values='rocket')
plt.title('Number of videos Published per Hour')
plt.xlable('Hour of day')
plt.ylable('No of videos')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=ad, x='views', y='likes')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```
